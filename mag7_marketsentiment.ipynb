```python
import yfinance as yf
import pandas as pd
import datetime as datetime

tickers = ["AAPL", "AMZN", "GOOG", "META", "MSFT", "NVDA", "TSLA"]

data = yf.download(tickers, interval="1d", auto_adjust=True, group_by="ticker")

# flatten to one DataFrame
rows = []
for t in tickers:
    df = data[t].reset_index()
    df["symbol"] = t
    rows.append(df)

prices = pd.concat(rows)
prices.head()
```

```python
prices.to_csv("mag7_prices.csv", index=False)
```

```python
rows = []

for t in tickers:
    df = data[t].reset_index()
    df["symbol"] = t
    rows.append(df)

prices = pd.concat(rows).sort_values(["symbol", "Date"])
prices.head()
```

```python
import feedparser

rss_url = "https://feeds.finance.yahoo.com/rss/2.0/headline?s=AAPL&region=US&lang=en-US"
feed = feedparser.parse(rss_url)

yahoo_headlines = [entry.title for entry in feed.entries]
yahoo_headlines[:10]
```

```python
rss_url = "https://www.marketwatch.com/rss/topstories"

feed = feedparser.parse(rss_url)

mw_headlines = [entry.title for entry in feed.entries]

df_marketwatch = pd.DataFrame({
    "source": "MarketWatch",
    "headline": mw_headlines,
    "date": datetime.date.today()
})

df_marketwatch.head(10)

```

```python
rss_cnbc = "https://www.cnbc.com/id/100003114/device/rss/rss.html"
feed_cnbc = feedparser.parse(rss_cnbc)

df_cnbc = pd.DataFrame({
    "source": "CNBC",
    "headline": [entry.title for entry in feed_cnbc.entries],
    "date": datetime.date.today()
})

df_cnbc.head(10)
```

```python
rss_investing = "https://www.investing.com/rss/news.rss"
feed_investing = feedparser.parse(rss_investing)

df_investing = pd.DataFrame({
    "source": "Investing.com",
    "headline": [entry.title for entry in feed_investing.entries],
    "date": datetime.date.today()
})

df_investing.head(10)
```

```python
import feedparser
import pandas as pd

def get_yahoo_finance_news(ticker="AAPL"):
    rss_url = f"https://feeds.finance.yahoo.com/rss/2.0/headline?s={ticker}&region=US&lang=en-US"
    feed = feedparser.parse(rss_url)

    news_list = []
    for entry in feed.entries:
        news_list.append({
            "ticker": ticker,
            "headline": entry.title,
            "link": entry.link,
            "published": entry.published
        })

    return pd.DataFrame(news_list)

# Example: Get MAG7 news
tickers = ["AAPL", "AMZN", "GOOG", "META", "MSFT", "NVDA", "TSLA"]

df_yahoo_list = [get_yahoo_finance_news(t) for t in tickers]
df_yahoo = pd.concat(df_yahoo_list, ignore_index=True)

df_yahoo.head(20)
```

```python
df_news = pd.concat(
    [
        df_yahoo,
        df_marketwatch,
        df_cnbc,
        df_investing
    ],
    ignore_index=True
)

df_news = df_news.drop_duplicates(subset=["headline"]).reset_index(drop=True)
df_news.head(20)
```

```python
df_news = pd.concat(
    [
        df_yahoo,
        df_marketwatch,
        df_cnbc,
        df_investing
    ],
    ignore_index=True
)

df_news = df_news.drop_duplicates(subset=["headline"]).reset_index(drop=True)
df_news.head(20)

df_news = df_news.rename(columns={
    "title": "headline",
    "summary": "description",
    "date": "published_at"
})

```

```python
df_news.to_csv("news_headlines_weekly.csv", index=False, encoding="utf-8")
```
